# İleri Birleştirme İşlemleri

```python
import numpy as np
import pandas as pd
```

```python
# birebir birleştirme
```

```python
df1 = pd.DataFrame({
    'calisanlar': ["Ali", "Veli", "Ayse", "Fatma"],
    'grup': ["Muhasebe", "Muhendislik", "Muhendislik", "İK"]
})
df1
```

```python
df2 = pd.DataFrame({
    'calisanlar': ["Ayse", "Ali", "Veli", "Fatma"],
    'ilk_giris': [2010, 2009, 2014, 2019]
})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="calisanlar")
```

```python
# Çoktan teke, many to one
```

```python
df3 = pd.merge(df1, df2)
df3
```

```python
df4 = pd.DataFrame({
    'grup': ["Muhasebe", "Muhendislik", "İK"],
    'mudur': ["Caner", "Mustafa", "Berkcan"]
})
df4
```

```python
pd.merge(df3, df4)
```

```python
# Çoktan Çoka, many to many
```

```python
df5 = pd.DataFrame({
    'grup': ["muhasebe", "Muhasebe", "Muhendislik", "Muhendislik", "İK", "İK"],
    'yetenekler': ["matematik", "excel", "kodlama", "linux", "excel", "yonetim"]
})
df5
```

```python
pd.merge(df1, df5)
```

```python
deneme1 = pd.DataFrame({
    "ogrenci": ["murat", "dogan", "soner"],
    "yas": [22, 21, 23]
})
deneme1
```

```python
deneme2 = pd.DataFrame({
    "ogrenci": ["murat", "soner", "dogan"],
    "cinsiyet": ["Erkek", "Erkek", "Erkek"]
})
deneme2
```

```python
pd.merge(deneme1, deneme2, on="ogrenci")
```

```python
?pd.merge
```

```python
join = pd.merge(deneme1, deneme2)
join
```

```python
deneme3 = pd.DataFrame({
    "ogrenci": ["murat", "murat", "dogan", "soner"],
    "aldigi_dersler": ["python", "yapay zeka", "tasarım", "3d tasarım"]
})
deneme3 
```

```python
pd.merge(join, deneme3)
```

```python
dictionary = {
    "ogrenci": ["murat", "soner", "havvanur", "ramazan"],
    "id": ["murat01", "soner01", "havvanur01", "ramazan01"]
}

dataframe = pd.DataFrame(dictionary)
dataframe
```

```python
dictionary2 = {
    "id": ["murat01", "soner01", "havvanur01", "ramazan01"],
    "ogretmen": ["gercel matematik", "senol hoca", "okan hoca", "murat hoca"]
}
dataframe2 = pd.DataFrame(dictionary2)
dataframe2
```

```python
pd.merge(dataframe, dataframe2, on="id")
```
