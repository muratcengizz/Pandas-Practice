# İleri Toplulaştırma İşlemleri (Aggregate, Filter, Transform, Apply)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
    'degisken1': [10,23, 33, 22, 11, 99],
    'degisken2': [100, 253, 333, 262, 111, 969]
}, columns=['gruplar', 'degisken1', 'degisken2'])
df
```

```python
# Aggregate
# Dışarıdan veya pandas içerisinden birden fazla fonksiyonu bir liste içerisinde çağırıp çalıştırabiliriz.
```

```python
df.groupby('gruplar').aggregate(['min', np.median, max])
```

```python
df.groupby('gruplar').aggregate([min, max])
```

```python
df.groupby('gruplar').aggregate(["describe", max])
```

```python
df.groupby('gruplar').aggregate({'degisken1':'min', 'degisken2':'max'})
```

```python
df.groupby('gruplar').aggregate({'degisken1':max, 'degisken2':min})
```

```python
df.groupby('gruplar').aggregate({'degisken1':np.median, 'degisken2':np.shape})
```
