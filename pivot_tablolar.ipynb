# Pivot Tablolar

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby('sex')[['fare']].sum()
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
titanic.groupby(['sex', 'class'])[['survived']].aggregate('mean').unstack()
```

```python
titanic.groupby(['sex', 'class'])[['survived']].aggregate('mean')
```

```python
titanic.groupby(['sex', 'class'])[['survived']].transform('mean')
```

```python
titanic.groupby(['sex', 'class'])[['survived']].aggregate(['mean', 'std'])
```

```python
titanic.groupby(['sex', 'class'])[['survived']].aggregate(['mean', 'std']).unstack()
```

# Pivot Table

```python
titanic.pivot_table('survived', index='sex', columns='class')
```

```python
titanic.pivot_table('survived', index='class', columns='sex')
```
