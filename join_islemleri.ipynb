```python
array1 = np.random.randint(1, 20, size=(3,3))
array2 = np.random.randint(1, 30, size=(5, 3))
```

```python
df1 = pd.DataFrame(array1)
df2 = pd.DataFrame(array2)
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
data1 = pd.concat([df1, df2], keys=["df1", "df2"])
data1
```

```python
data2 = pd.concat([df1, df2], axis=1, join="inner")
data2
```

```python
data3 = pd.concat([df1, df2], axis=1, join="outer")
data3
```

```python
data3.columns = ["kolon1", "kolon2", "kolon3", "kolon4", "kolon5", "kolon6"]
```
