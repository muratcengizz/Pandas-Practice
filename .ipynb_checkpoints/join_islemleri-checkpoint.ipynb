```python
import numpy as np
import pandas as pd
```

```python
m = np.random.randint(1, 30, size=(5,3))
df1 = pd.DataFrame(m, columns=["var1", "var2", "var3"])
df1
```

```python
df2 = df1 + 99
df2
```

```python
?pd.concat
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
deneme1 = pd.DataFrame(np.random.randint(1,20,size=(3,2)), columns=["kolon1", "kolon2"], index=["index1", "index2", "index3"])
deneme2 = pd.DataFrame(np.random.randint(39, 56, size=(2,2)))
print(deneme1, deneme2)
```

```python
pd.concat([deneme1,deneme2], ignore_index=False)
```

```python
pd.concat([deneme1, deneme2], ignore_index=True)
```
